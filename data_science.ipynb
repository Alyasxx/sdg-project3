```python
#Import Nominatim
#Nominatim digunakan untuk mencari data pada OpenStreetMap berdasar alamat atau lokasi
#Install Geopy dulu
```

```python
pip install geopy
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="app" , timeout=None)
```

```python
lat = [] #List menyimpan data latitude
lon = [] #List menyimpan data longitude

for name in rest_loc['Name']:
    location = geolocator.geocode(name)
    
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
        
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
print(lat)
```

```python
print(lon)
```

```python
#Membuat kolom baru dalam df
#Deklarasikan dulu kolomnya
```

```python
rest_loc['lat'] = lat
rest_loc['lon'] = lon
```
